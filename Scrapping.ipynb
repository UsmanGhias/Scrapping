```python
# Text preprocessing

import csv
import re

# Function to clean the content using regular expressions
def clean_content(content):
    # Remove extra spaces
    content = re.sub(r'\s+', ' ', content)

    # Remove special characters (retain only alphanumeric and space)
    content = re.sub(r'[^a-zA-Z0-9\s]', '', content)

    # Remove broken links (assuming links start with 'http' or 'www')
    content = re.sub(r'http\S+|www\S+', '', content)

    return content

# Specify the CSV file name
csv_file = 'corpus_data_cleaned.csv'

# Open the CSV file in write mode
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    # Create a CSV writer
    csv_writer = csv.writer(file)

    # Write the header row
    csv_writer.writerow(['Topic', 'Corpus'])

    # Write the topic names and cleaned corpus into the CSV file
    for topic, corpus in corpus_dict.items():
        cleaned_corpus = clean_content(corpus)
        csv_writer.writerow([topic, cleaned_corpus])

print(f"Cleaned corpus data has been stored in '{csv_file}'.")
```

```python
import csv
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords

# Assuming you have the 'topics' list and 'corpus_dict' dictionary from the previous code

# Function to calculate stems and lemmas
def calculate_tokens(content):
    tokens = word_tokenize(content)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token.lower() for token in tokens if token.isalnum() and token.lower() not in stop_words]

    # Perform stemming
    stemmer = PorterStemmer()
    stems = [stemmer.stem(token) for token in tokens]

    # Perform lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(token) for token in tokens]

    return stems, lemmas

# Specify the CSV file name
csv_file = 'tokens.csv'

# Open the CSV file in write mode
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    # Create a CSV writer
    csv_writer = csv.writer(file)

    # Write the header row
    csv_writer.writerow(['Topic', 'Stems', 'Lemmas'])

    # Write the data into the CSV file
    for topic in topics:
        content = corpus_dict.get(topic, "")

        # Calculate stems and lemmas
        stems, lemmas = calculate_tokens(content)

        # Write a row for each topic
        csv_writer.writerow([topic, stems, lemmas])

print(f"Tokens data has been stored in '{csv_file}'.")
```

```python
import csv
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Function to calculate unique words
def calculate_unique_words(content):
    tokens = word_tokenize(content)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token.lower() for token in tokens if token.isalnum() and token.lower() not in stop_words]

    # Calculate unique words
    unique_words = set(tokens)

    return unique_words

# Specify the CSV file name
csv_file = 'vocabulary.csv'

# Open the CSV file in write mode
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    # Create a CSV writer
    csv_writer = csv.writer(file)

    # Write the header row
    csv_writer.writerow(['Topic', 'Unique Words'])

    # Write the data into the CSV file
    for topic in topics:
        content = corpus_dict.get(topic, "")

        # Calculate unique words
        unique_words = calculate_unique_words(content)

        # Write a row for each topic
        csv_writer.writerow([topic, unique_words])

print(f"Vocabulary data has been stored in '{csv_file}'.")
```

```python
import csv
from nltk import pos_tag, word_tokenize
from nltk.corpus import stopwords


nltk.download('averaged_perceptron_tagger')

# Function to perform POS tagging
def pos_tagging(content):
    tokens = word_tokenize(content)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token.lower() for token in tokens if token.isalnum() and token.lower() not in stop_words]

    # Perform POS tagging
    pos_tags = pos_tag(tokens)

    return pos_tags

# Specify the CSV file name
csv_file = 'pos.csv'

# Open the CSV file in write mode
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    # Create a CSV writer
    csv_writer = csv.writer(file)

    # Write the header row
    csv_writer.writerow(['Topic', 'Data with POS Tags'])

    # Write the data into the CSV file
    for topic in topics:
        content = corpus_dict.get(topic, "")

        # Perform POS tagging
        pos_tags = pos_tagging(content)

        # Convert pos_tags to a string for CSV
        pos_tags_str = ' '.join([f"{word}_{pos}" for word, pos in pos_tags])

        # Write a row for each topic
        csv_writer.writerow([topic, pos_tags_str])

print(f"POS data has been stored in '{csv_file}'.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Load the data from readability.csv
readability_data = pd.read_csv('readability.csv')

# Check unique values in 'Readability' column
unique_values = readability_data['Readability'].unique()
print("Unique Values in 'Readability' Column:", unique_values)

# Raise an error if there is only one unique class
if len(unique_values) < 2:
    raise ValueError("There should be at least two unique classes in the 'Readability' column.")

# Convert readability levels to numerical labels
readability_data['Label'] = readability_data['Readability'].apply(lambda x: 0 if x == 'easy' else 1)

# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(
    readability_data['Topic'], readability_data['Label'], test_size=0.2, random_state=42
)

# Check unique classes in training labels
unique_classes_train = train_labels.unique()
if len(unique_classes_train) < 2:
    raise ValueError("There should be at least two unique classes in the training labels.")

# Convert text data to numerical features using CountVectorizer
vectorizer = CountVectorizer()
train_features = vectorizer.fit_transform(train_data)
test_features = vectorizer.transform(test_data)

# Train a logistic regression model
model = LogisticRegression()
model.fit(train_features, train_labels)

# Make predictions on the testing set
predictions = model.predict(test_features)

# Evaluate the model
accuracy = accuracy_score(test_labels, predictions)
classification_report_str = classification_report(test_labels, predictions)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report_str)
```
